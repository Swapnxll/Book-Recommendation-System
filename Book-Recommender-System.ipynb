```python
##popularity based recommender system
##consider those books which has highest avg rating 
## with condition min 250 reviews
```

```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df= num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df.info()
```

```python
new_df=popular_df[popular_df['num_ratings'] >= 250].sort_values(by='avg_rating', ascending=False).head(50)
new_df
```

```python
merged_df = new_df.merge(books, on='Book-Title', suffixes=('_pop', '_book'))
popular_df=merged_df
```

```python
popular_df=popular_df.drop_duplicates('Book-Title')
```

```python
popular_df
```

```python
## COlABORATIVE FILTERING BASED RECOMMENDER SYSTEM
##user who have given more than 200 ratings
##books which have more than 50 reviews
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
## now we can imagine one book as 810 dimensional vector 
##and based on the Euclidean distance between each vector we can find similarity between them
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
##this shows similarity score of every book with itself and other books
```

```python
##this shows the similarity of book 1984 [index of 1984 is 0]with every other.with itself it is 1
sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1],reverse=True)[1:6] ##[index of book which is being compared to 1984, score]
```

```python
def recommend(book_name):
    #index fetch
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        
        temp_df=(books[books['Book-Title']==pt.index[i[0]]])
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```

```python
recommend('1984')
```
